### Metric

Because of the long tail of distribution and multi modal form of that,I decided to pick percentile or Median metrics instead of mean. 
<br>The metrics should be **robust** and **sensitive**

#### Check *robustness* of metric
For checking robustness of metric,I randomly sampled from the control group and calculated the metrics for that subsample.<br>The robust metrics should not zigzagging alot.

```python
#check robustness of metric
median,perc_60,perc_65,perc_70,perc_75,perc_80,perc_85,perc_90 = [],[],[],[],[],[],[],[]
for x in range(10):
    temp = control_g.sample(frac=0.15)
    median.append(temp.diff_time.median())
    perc_60.append(temp.diff_time.quantile(0.6))
    perc_65.append(temp.diff_time.quantile(0.65))
    perc_70.append(temp.diff_time.quantile(0.7))
    perc_75.append(temp.diff_time.quantile(0.75))
    perc_80.append(temp.diff_time.quantile(0.80))
    perc_85.append(temp.diff_time.quantile(0.85))
    perc_90.append(temp.diff_time.quantile(0.9))
    

test = pd.DataFrame([median,perc_60,perc_65,perc_70,perc_75,perc_80,perc_85,perc_90]).T
test.columns = ['median','perc_60','perc_65','perc_70','perc_75','perc_80','perc_85','perc_90']
test.plot()

```

*Base on plot median,65% and 70% percentile are robust metrics*

#### Check *sensitivity* of metric
To checking if the metrics are sensitive to changes,I calculate those in different hours of day, and see the changes. If the metric do not change, It shows that the metric is not enough sensitive. Better way to check the sensitivity is to compare some rainy days or the day with promotions with normal day, and check if the metric are sensitive or not.

```python
#check sensitivity of metric
gp = control_g.groupby(control_g.Order_time.dt.hour)

median,perc_60,perc_65,perc_70,perc_75,perc_80,perc_85,perc_90 = [],[],[],[],[],[],[],[]
for x in range(10):
    median = gp.diff_time.median()
    perc_60 = gp.diff_time.quantile(0.6)
    perc_65 = gp.diff_time.quantile(0.65)
    perc_70 = gp.diff_time.quantile(0.7)
    perc_75 = gp.diff_time.quantile(0.75)
    perc_80 = gp.diff_time.quantile(0.80)
    perc_85 = gp.diff_time.quantile(0.85)
    perc_90 = gp.diff_time.quantile(0.9)
    

test = pd.DataFrame([median,perc_60,perc_65,perc_70,perc_75,perc_80,perc_85,perc_90]).T
test.columns = ['median','perc_60','perc_65','perc_70','perc_75','perc_80','perc_85','perc_90']
test.plot()
```

*Base on this plot the median is more sensitive to changes compare to 65% and 70% percentile*

### Modeling
I used the Bayesian model to the priors that can generate this form of data. After that samples from posteriors(generated from different priors) and find distribution for median for both groups and compare two distributions to find the impact of the new algorithm.

I modeled the data with gaussian mixture model(4 normal) and Dirichlet distribution as there weighings.
<br>For finding the parameters, instead of using McMC methods, I used Automatic Differentiation Variational Inference(ADVL).
<br>I checked prior sensivity but didn't attached it to this notebook.

```python
with pm.Model() as ab_test:
    w_a = pm.Dirichlet('w_a', np.array([1, 1, 1,1]))
    mu_a = pm.Uniform('mu_a', 0., 120., shape=4)
    sigma_a = pm.Uniform('sigma_a',0,50,shape=4)
    ##
    likelihood_a = pm.NormalMixture('likelihood_a', w_a, mu_a, sd=sigma_a, observed=control_g.diff_time)
    #############################################
    w_b = pm.Dirichlet('w_b', np.array([1, 1, 1,1]))
    mu_b = pm.Uniform('mu_b', 0., 120., shape=4)
    sigma_b = pm.Uniform('sigma_b',0,50,shape=4)
    ##
    likelihood_b = pm.NormalMixture('likelihood_b', w_b, mu_b, sd=sigma_b, observed=treat_g.diff_time)
    #############################################

with ab_test:
    ab_trace = pm.fit(random_seed=RANDOM_SEED,n=35000).sample(6000)
```

```python
with ab_test:
    ppc_trace = pm.sample_ppc(ab_trace, 5000, random_seed=RANDOM_SEED)
```

After fiting the model and sample from posteriors,I tried some sanity check to check if the models are good.

**Sanity check** of the posterior result compare to actual data for control group:

```python
#sanity check of the posterior result compare to actual data for control group
f, ax = plt.subplots(1, 1)
sns.distplot(ppc_trace['likelihood_a'].flatten(),color = 'red',label = 'Control group posterior')
sns.distplot(control_g.diff_time,label='True data')
plt.legend()
```

```python
#sanity check of the posterior result compare to actual data for treat group
sns.distplot(ppc_trace['likelihood_b'].flatten(),color='red',label = 'treat group osterior')
sns.distplot(treat_g.diff_time,label='True data')
plt.legend()
```

Compare two distributions:

```python
b_less_a = np.mean(median_b<median_a)
sns.distplot((median_b- median_a),kde=False,norm_hist=True)
plt.plot([0]*2,np.linspace(0,0.08,2))
plt.text(-4, 0.08,'mean = {0:.3f}'.format(np.mean(median_b-median_a)))
plt.text(-23, 0.05,'{0:.3f} < meadian(b) - median(a)< {1:.3f}'.format(b_less_a,1-b_less_a))

#This plot shows that there is 89.1% chance that median of treat group is less than control group
#and we expect the median of treat group be 7.47 seconds better than control group
```
